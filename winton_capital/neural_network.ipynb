```python
sess = tf.InteractiveSession()
```

```python
feature_start = 26
feature_end = 147
truth_start = 147
truth_end = 209
feature_num = feature_end - feature_start
out_num = truth_end - truth_start
hidden = 1000
```

```python
x = tf.placeholder(tf.float32, shape=[None, feature_num])
y_ = tf.placeholder(tf.float32, shape=[None, out_num])
l2 = tf.placeholder(tf.float32)
```

```python
with tf.name_scope('hidden'):
    weights = tf.Variable(
        tf.truncated_normal([feature_num, hidden],
                            stddev=1.0/np.sqrt(float(feature_num)),
                            dtype=tf.float32),
        name='weights')
    biases = tf.Variable(tf.zeros([hidden], dtype=tf.float32),
                         name='biases')
    hidden1 = tf.nn.relu(tf.matmul(x, weights) + biases)
    reg_hidden1 = tf.nn.l2_loss(weights) + tf.nn.l2_loss(biases)
```

```python
with tf.name_scope('output'):
    weights = tf.Variable(
        tf.truncated_normal([hidden, out_num],
                            stddev=1.0/np.sqrt(float(hidden)),
                            dtype=tf.float32),
        name='weights')
    biases = tf.Variable(tf.zeros([out_num], dtype=tf.float32),
                         name='biases')
    logits = tf.matmul(hidden1, weights) + biases
    reg_hidden2 = tf.nn.l2_loss(weights) + tf.nn.l2_loss(biases)
```

```python
sess.run(tf.initialize_all_variables())
```

```python
y = logits
```

```python
cost = tf.reduce_mean(tf.pow((y_ - y), 2)) + l2 * (reg_hidden1 + reg_hidden2)
```

```python
train_step = tf.train.GradientDescentOptimizer(.01).minimize(cost)
```
